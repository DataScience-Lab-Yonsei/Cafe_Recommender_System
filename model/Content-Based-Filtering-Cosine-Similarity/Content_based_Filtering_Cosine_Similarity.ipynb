## 1. Binary

### 키워드 개수를 binary 값으로 변경

```python
naver = naver_df[['"커피가 맛있어요"', '"친절해요"',
       '"음료가 맛있어요"', '"디저트가 맛있어요"', '"매장이 청결해요"', '"대화하기 좋아요"', '"집중하기 좋아요"',
       '"인테리어가 멋져요"', '"좌석이 편해요"', '"화장실이 깨끗해요"', '"특별한 메뉴가 있어요"',
       '"가성비가 좋아요"', '"뷰가 좋아요"', '"주차하기 편해요"', '"사진이 잘 나와요"']]
```

```python
naver.head()
```

```python
a = naver.astype(bool).astype(int)
a
```

### 카페 수와 카페 특징 수에 맞는 행렬 생성

```python
a.iloc[0].to_list()
list(a.iloc[0])
```

```python
score = np.zeros((7661, 15))
score.shape

for i in range(0,7661):
    score[i] = a.iloc[i].to_list()
    list(a.iloc[i])
```

```python
score[0]
```

```python
score[1]
```

```python
score[7660]
```

```python
print(score.shape, end='\n\n')
print(score)
```

### 코사인 유사도 계산

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similar = cosine_similarity(score, score)

for i in range(7661):
    cosine_similar[i,i] = 1.0

cosine_similar_data = pd.DataFrame(cosine_similar,
                   index= naver_df['cafe'],
                   columns=naver_df['cafe'])
cosine_similar_data.head(10)

print(cosine_similar_data)
```

### 추천 함수 정의

```python
def content_recommender(name, n_of_recomm):
    # 주어진 카페와 다른 카페의 특징의 similarity를 가져온다
    sim_scores = cosine_similar_data[name]
    # similarity 기준으로 정렬하고 n_of_recomm만큼 가져오기 (자기자신은 빼기)
    sim_scores = sim_scores.sort_values(ascending=False)[1:n_of_recomm+1]
    # 비슷한 카페의 index number 반환
    return sim_scores
```

```python
print(content_recommender('버블베어 방화동', 5))
```

#### 거리지표 추가

```python
public_df = pd.read_csv('C:\\Users\\jaewo\\Downloads\\publicdata_cafe_lastdrop.csv', index_col=0)
```

```python
public_df.head()
```

```python
def content_recommender2(name, n_of_recomm, address):
    # 주어진 카페와 다른 카페의 특징의 similarity를 가져온다
    sim_scores = cosine_similar_data[name]
    # 비슷한 카페의 index number 반환
    sim_scores = sim_scores.sort_values(ascending=False)
    recom_cafe = pd.DataFrame(sim_scores, index=sim_scores.index)
    # 거리지표 추가
    recommendations = pd.merge(recom_cafe, public_df.set_index('cafe')[['시군구명','도로명주소']], left_index=True,right_index=True,how='left')
    # address 부분
    recommendations = recommendations[recommendations['시군구명']==address]
    R = recommendations.sort_values(by=[name], axis=0, ascending = False)[1:n_of_recomm+1]
    recommendations = R.iloc[:n_of_recomm,:]
    recommendations.columns = ['유사도','시군구명','도로명주소']
    return recommendations
```

```python
content_recommender2('메가커피 연희점 연희동', 3, '강남구')
```

```python
content_recommender2('안다르커피 창천동', 10, '서대문구')
```
