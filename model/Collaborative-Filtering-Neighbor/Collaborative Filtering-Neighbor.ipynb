### 함수 정의 (RMSE, 코사인 유사도, 예측치 계산)

```python
# train set 사용자들의 Cosine similarities 계산
from sklearn.metrics.pairwise import cosine_similarity
matrix_dummy = rating_matrix.copy().fillna(0)
user_similarity = cosine_similarity(matrix_dummy, matrix_dummy)
user_similarity = pd.DataFrame(user_similarity, index=rating_matrix.index, columns=rating_matrix.index)
```

```python
# 정확도(RMSE)를 계산하는 함수 
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))

# 모델별 RMSE를 계산하는 함수 
def score(model, neighbor_size=0):
    id_pairs = zip(x_test['user_id'], x_test['cafe'])
    y_pred = np.array([model(user, cafe, neighbor_size) for (user, cafe) in id_pairs])
    y_true = np.array(x_test['rating'])
    return RMSE(y_true, y_pred)
```

```python
# Neighbor size를 정해서 예측치를 계산하는 함수 
def cf_knn(user_id, cafe, neighbor_size=0):
    if cafe in rating_matrix:
        # 현재 사용자와 다른 사용자 간의 similarity 가져오기
        sim_scores = user_similarity[user_id].copy()
        # 현재 카페에 대한 모든 사용자의 rating값 가져오기
        cafe_ratings = rating_matrix[cafe].copy()
        # 현재 카페를 평가하지 않은 사용자의 index 가져오기
        none_rating_idx = cafe_ratings[cafe_ratings.isnull()].index
        # 현재 카페를 평가하지 않은 사용자의 rating (null) 제거
        cafe_ratings = cafe_ratings.drop(none_rating_idx)
        # 현재 카페를 평가하지 않은 사용자의 similarity값 제거
        sim_scores = sim_scores.drop(none_rating_idx)
##### Neighbor size가 지정되지 않은 경우        
        if neighbor_size == 0:          
            # 현재 카페를 평가한 모든 사용자의 가중평균값 구하기
            mean_rating = np.dot(sim_scores, cafe_ratings) / sim_scores.sum()
##### Neighbor size가 지정된 경우
        else:                       
            # 해당 카페를 평가한 사용자가 최소 2명이 되는 경우에만 계산
            if len(sim_scores) > 1: 
                # 지정된 neighbor size 값과 해당 카페를 평가한 총사용자 수 중 작은 것으로 결정
                neighbor_size = min(neighbor_size, len(sim_scores))
                # array로 바꾸기 (argsort를 사용하기 위함)
                sim_scores = np.array(sim_scores)
                cafe_ratings = np.array(cafe_ratings)
                # 유사도를 순서대로 정렬
                user_idx = np.argsort(sim_scores)
                # 유사도를 neighbor size만큼 받기
                sim_scores = sim_scores[user_idx][-neighbor_size:]
                # 카페 rating을 neighbor size만큼 받기
                cafe_ratings = cafe_ratings[user_idx][-neighbor_size:]
                # 최종 예측값 계산 
                mean_rating = np.dot(sim_scores, cafe_ratings) / sim_scores.sum()
            else:
                mean_rating = 3.0
    else:
        mean_rating = 3.0
    return mean_rating
```

```python
# 정확도 계산
score(cf_knn, neighbor_size=30)
```

```python
# train 데이터의 user의 rating 평균과 카페의 평점편차 계산 
rating_mean = rating_matrix.mean(axis=1)
rating_bias = (rating_matrix.T - rating_mean).T

def CF_knn_bias(user_id, cafe, neighbor_size=0):
    if cafe in rating_bias:
        # 현 user와 다른 사용자 간의 유사도 가져오기
        sim_scores = user_similarity[user_id].copy()
        # 현 cafe의 평점편차 가져오기
        cafe_ratings = rating_bias[cafe].copy()
        # 현 cafe에 대한 rating이 없는 사용자 삭제
        none_rating_idx = cafe_ratings[cafe_ratings.isnull()].index
        cafe_ratings = cafe_ratings.drop(none_rating_idx)
        sim_scores = sim_scores.drop(none_rating_idx)
        
##### Neighbor size가 지정되지 않은 경우        
        if neighbor_size == 0:
            # 편차로 예측값(편차 예측값) 계산
            prediction = np.dot(sim_scores, cafe_ratings) / sim_scores.sum()
            # 편차 예측값에 현 사용자의 평균 더하기
            prediction = prediction + rating_mean[user_id]
##### Neighbor size가 지정된 경우            
        else:
            # 해당 카페를 평가한 사용자가 최소 2명이 되는 경우에만 계산            
            if len(sim_scores) > 1:
                # 지정된 neighbor size 값과 해당 카페를 평가한 총사용자 수 중 작은 것으로 결정
                neighbor_size = min(neighbor_size, len(sim_scores))
                # array로 바꾸기 (argsort를 사용하기 위함)
                sim_scores = np.array(sim_scores)
                cafe_ratings = np.array(cafe_ratings)
                # 유사도를 순서대로 정렬
                user_idx = np.argsort(sim_scores)
                # 유사도와 rating을 neighbor size만큼 받기
                sim_scores = sim_scores[user_idx][-neighbor_size:]
                cafe_ratings = cafe_ratings[user_idx][-neighbor_size:]
                # 편차로 예측치 계산
                prediction = np.dot(sim_scores, cafe_ratings) / sim_scores.sum()
                # 예측값에 현 사용자의 평균 더하기
                prediction = prediction + rating_mean[user_id]
            else:
                prediction = rating_mean[user_id]
    else:
        prediction = rating_mean[user_id]
    return prediction
```

### 아이템 기반 CF

```python
# train set의 모든 가능한 아이템 pair의 Cosine similarities 계산
from sklearn.metrics.pairwise import cosine_similarity
rating_matrix_t = np.transpose(rating_matrix)
matrix_dummy = rating_matrix_t.copy().fillna(0)
item_similarity = cosine_similarity(matrix_dummy, matrix_dummy)
item_similarity = pd.DataFrame(item_similarity, index=rating_matrix_t.index, columns=rating_matrix_t.index)

# 주어진 카페의 가중평균 rating을 계산하는 함수, 
# 가중치는 주어진 아이템과 다른 아이템 간의 유사도(item_similarity)
def CF_IBCF(user_id, cafe, user_rating):
    if cafe in item_similarity:      # 현재 카페가 train set에 있는지 확인
        # 현재 카페와 다른 카페의 similarity 값 가져오기
        sim_scores = item_similarity[cafe]
        # 현 사용자의 모든 rating 값 가져오기
        user_rating = rating_matrix_t[user_id]
        # 사용자가 평가하지 않은 카페 index 가져오기
        non_rating_idx = user_rating[user_rating.isnull()].index
        # 사용자가 평가하지 않은 카페 제거
        user_rating = user_rating.dropna()
        # 사용자가 평가하지 않은 카페의 similarity 값 제거
        sim_scores = sim_scores.drop(non_rating_idx)
        # 현 카페에 대한 예상 rating 계산, 가중치는 현 카페와 사용자가 평가한 카페의 유사도
        mean_rating = np.dot(sim_scores, user_rating) / sim_scores.sum()
    else:
        mean_rating = 3.0
    return mean_rating
```
