# 네이버

```python
df = pd.read_csv('data/public_data/publicdata_cafe_lastdrop.csv')
```

```python
naver_review = pd.read_csv('data/naver_data/naver_review_name.csv')
naver_review.head()
```

```python
print(len(naver_review))
```

```python
naver_more = pd.merge(naver_review, df[['cafe','상권업종소분류명']], how='inner', on='cafe')
```

```python
len(naver_more)
```

```python
naver_more['상권업종소분류명'].unique()
```

```python
naver_more.groupby('상권업종소분류명').count()
```

```python
naver_more['사주카페'] = naver_more['상권업종소분류명']=='사주카페'
naver_more['생과일주스전문점'] = naver_more['상권업종소분류명']=='생과일주스전문점'
naver_more['애견카페'] = naver_more['상권업종소분류명']=='애견카페'
naver_more['전통찻집/인삼찻집'] = naver_more['상권업종소분류명']=='전통찻집/인삼찻집'
naver_more['커피전문점/카페/다방'] = naver_more['상권업종소분류명']=='커피전문점/카페/다방'
```

```python
naver_more.head()
```

```python
kakao_review = pd.read_csv("/data/kakao_data/ratings.csv")
print(len(kakao_review))
```

```python
kakao_review.columns
```

```python
pd.merge(naver_more, kakao_review[['cafe','user_id','rating']], how='inner', on='cafe').drop(['search_name','address','road_address','near_station','station_distance'],axis=1)
```

```python
kakao_naver_review = kakao_review.merge(naver_review, how='inner',on='cafe')
```

```python
kakao_naver_review['cafe'].nunique()
```

```python
print(kakao_naver_review)
```

```python
kakao_naver_review.to_csv('/model/Content-Based-Filtering-ML/kakao_naver_review.csv',index=False)
```
