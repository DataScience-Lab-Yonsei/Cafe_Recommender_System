# reviews_three

```python
df = pd.read_csv('/content/drive/MyDrive/DSL/[프로젝트] 모델링/A3NCF/topic model/data/A3NCF_topicmodel_input_df_3.csv')
df.head()
```

```python
# 전체 데이터셋 길이 확인
len(df)
```

```python
# train, test, valid 데이터셋 길이 예상
train_num = round(len(df) * 0.8)
test_num = round(len(df) * 0.1)
valid_num = round(len(df) * 0.1)
print(f"length of train data set     : {train_num}")
print(f"length of test data set      : {test_num}")
print(f"length of validation data set: {valid_num}")
```

```python
# 무작위 분할
df_train = df.sample(frac=0.8, random_state=1215)
df_remain = df.drop(df_train.index)
df_test = df_remain.sample(frac=0.5, random_state=1215)
df_valid = df_remain.drop(df_test.index)
```

```python
# train, test, valid 데이터셋 길이 확인
df_train_num = len(df_train)
df_test_num = len(df_test)
df_valid_num = len(df_valid)
print(f"length of train data set     : {df_train_num}")
print(f"length of test data set      : {df_test_num}")
print(f"length of validation data set: {df_valid_num}")
```

```python
# train, test, valid 데이터셋의 인덱스 확인
print(df_train.index.sort_values())
print(df_test.index.sort_values())
print(df_valid.index.sort_values())
```

```python
# 분할한 데이터셋 각각 저장
df_train.to_csv('/content/drive/MyDrive/DSL/[프로젝트] 모델링/A3NCF/topic model/data/A3NCF_train_3.csv',index=False)
df_test.to_csv('/content/drive/MyDrive/DSL/[프로젝트] 모델링/A3NCF/topic model/data/A3NCF_test_3.csv',index=False)
df_valid.to_csv('/content/drive/MyDrive/DSL/[프로젝트] 모델링/A3NCF/topic model/data/A3NCF_valid_3.csv',index=False)
```
