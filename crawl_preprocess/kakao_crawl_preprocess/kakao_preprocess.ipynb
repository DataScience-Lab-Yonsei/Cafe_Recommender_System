## pkl 합치기

```python
pkl_list = sorted(glob.glob('*.pkl'))
pkl_list
```

```python
pkl_list = sorted(glob.glob('*.pkl'))
zeroreview = []
nocafe = []
address_diff = []

for pkl in pkl_list:
  with open(pkl,'rb') as f:
    new_list = pickle.load(f)
    zeroreview = zeroreview + new_list['zeroreview']
    nocafe = nocafe + new_list['nocafe']
    address_diff = address_diff + new_list['address_diff']
  
  print(len(new_list['zeroreview']), '개 추가! 총', len(zeroreview), '개의 리뷰가 0개인 카페들이 존재합니다.')
  print(len(new_list['nocafe']), '개 추가! 총', len(nocafe), '개의 검색결과가 없는 카페들이 존재합니다.')
  print(len(new_list['address_diff']), '개 추가! 총', len(address_diff), '개의 검색결과는 있었지만 주소가 달라서 찾지 못한 카페들이 존재합니다.')
```

```python
print('한 번 더 확인하면 좋을 거 같은 애들 개수')
print(len(nocafe)+len(address_diff))
```

```python
indexdict = dict({'zeroreview': zeroreview,
     'nocafe':nocafe,
     'address_diff':address_diff})
```

```python
with open('index_notcrawled_total.pkl','wb') as f:
    pickle.dump(indexdict,f)
```

```python
filtering = df_rating.groupby(['user_id','cafe'], as_index=False).agg({'rating':['count']})[['user_id','cafe']]
```
