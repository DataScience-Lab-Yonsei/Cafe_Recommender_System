```python
df_findduplicate = pd.DataFrame()
for i in range(len(df)):
  namesame = df[df['cafe']==df['cafe'][i]]
  df_findduplicate = df_findduplicate.append(namesame.head(1), ignore_index=True)
```

```python
strange=[]
unique = df_findduplicate['cafe'].unique()
for i in range(len(unique)):
  a = len(df_findduplicate[df_findduplicate['cafe']==unique[i]])
  if a != 1:
    print(unique[i],':',a,'개의 매장 이름이 중복입니다. 확인이 필요합니다.')
    strange.append(unique[i])
print(len(strange))
```

```python
indexx = []
for i in range(len(strange)):
  ha = df[df['cafe']==strange[i]].index
  if len(ha)==2:
    indexx.append(ha[0])
  elif len(ha)==3:
    indexx.append(ha[0])
    indexx.append(ha[1])
  else:
    indexx.append(ha[0])
    indexx.append(ha[1])
    indexx.append(ha[2])
  print(len(indexx))
```
